### Modelização em Python
Os colares são ambos representados numa matriz com os numeros 0, 1, 2 e 3 em posições aleatórias da mesma. Cada Matriz é gerada aleatoriamente pela função geradorAleatorio() que recebe um dicionário em que cada posição do mesmo(0, 1, 2 ou 3, que representam as diferentes cores) contém quantas bolas dessa cor existem no problema.

Em cada um dos vetores, a posição 0 de cada linha da matriz representa a bola mais mais alta de cada colar e as bolas estão ordendas no sentido dos ponteiros do relógio.

Existem posições em comum em ambos os colares. As posições 3 e 7 do colar da esquerda são iguais, respetivamente, às posições 17 e 13 do colar da direita.  

As ações possiveis de exercer nos colares são:
'LL' - rotação colar da esquerda para a esquerda;
'LR' - rodar colar da esquerda para direira;
'RL' - rodar colar da direita para a esquerda;
'RR' - rodar colar da direita para a direita.

A cada rotação para a direita os elementos do respetivo colar que a sofre vão todos para a posição seguinte do vetor, sendo que o último elemento vai para a primeira posiçaõ do vetor; a cada rotação para a esquerda os elemtos do respetivo colar que a sofre vão todos uma posição para a esquerda no vetor, sendo que o primeiro elemento vai para a última posição do vetor.

Quando se realiza uma rotação no colar da esquerda são atualizadas as posições que representam as bolas em comum no outro colar e vice-versa.

Cada rotação é de custo unitário.

Alcança-se o objetivo quando em cada colar existem apenas duas cores, cada uma numa sequência sem interrupções.

```python
from searchPlus import *
import random

class PuzzleColares(Problem):
    """Engarrafamento"""


    def __init__(self, initial=None, goal=None):
        """Construtor: alterem os parâmetros a vosso gosto"""
        #As posições iguais nos dois colares são a 0 e 16
        self.initial = initial        
    
    def result(self, state, action): 
        """Do not forget the result"""
        colar1 = [None] * 20
        colar2 = [None] * 20
        stateAux = [colar1, colar2]
        if(action == "LR"):
            for i in range(20):
                stateAux[0][i] = state[0][i-1]
                stateAux[1][i] = state[1][i]
            stateAux[1][17] = stateAux[0][3]
            stateAux[1][13] = stateAux[0][7]
            return stateAux
        elif(action == "LL"):
            for i in range(20):
                stateAux[0][i-1] = state[0][i]
                stateAux[1][i] = state[1][i]
            stateAux[1][17] = stateAux[0][3]
            stateAux[1][13] = stateAux[0][7]
            return stateAux
        elif(action == "RL"):
            for i in range(20):
                stateAux[1][i-1] = state[1][i]
                stateAux[0][i] = state[0][i]
            stateAux[0][3] = stateAux[1][17]
            stateAux[0][7] = stateAux[1][13]
            return stateAux
        elif(action == "RR"):
            for i in range(20):
                stateAux[1][i] = state[1][i-1]
                stateAux[0][i] = state[0][i]
            stateAux[0][3] = stateAux[1][17]
            stateAux[0][7] = stateAux[1][13]
            return stateAux
        else:
            return stateAux
        
    def actions(self, state):
        """ The actions please"""
        return ("LL","LR","RL","RR")
    
    def goal_test(self,state):
        """ Goal test please"""
        if sequencia(state[1],-2,6) and sequencia(state[1], 7,16) and sequencia(state[0], -3, 6) and sequencia(state[0],8,16):
            return True
        elif sequencia(state[1],4,12) and sequencia(state[1], -6,3) and sequencia(state[0], 4, 13) and sequencia(state[0],-6,2):
            return True
        return False
        
    def path_cost(self, cost,state,action,new_state):
        """ compute the cost of a transition"""
        return cost+1
    
    def display(self, state):
        """ print the state please"""
        for i in range(11):
            if i == 0 or i == 10:
                print("     " + str(state[0][i]) + "     " + str(state[1][i]))
            elif i == 1 or i ==9:
                print("    " + str(state[0][-i]) + " " + str(state[0][i]) + "   " + str(state[1][-i]) + " " + str(state[1][i]))
            elif i == 2 or i == 8:
                print("   " + str(state[0][-i]) + "   " + str(state[0][i]) + " " + str(state[1][-i]) + "   " + str(state[1][i]))
            elif i == 3 or i == 7:
                print("  " + str(state[0][-i]) + "     " + str(state[1][-i]) + "     " + str(state[1][i]))
            elif i == 4 or i == 6:
                print(" " + str(state[0][-i]) + "     " + str(state[1][-i]) + " "+ str(state[0][i]) + "     " + str(state[1][i]))
            elif i == 5:
                print(str(state[0][-i]) + "     " + str(state[1][-i]) + "   " + str(state[0][i]) + "     " + str(state[1][i]))
        
def sequencia(state, indexI, indexF):
    temp = state[indexI]
    for i in range(indexI, indexF+1):
        if state[i] != temp:
            return False
    return True
        
def exec(p,estado,accoes):
    """ Executa uma sequência de acções a partir do estado
        devolve um par (estado, custo) depois de imprimir
    """
    custo = 0
    for a in accoes:
        seg = p.result(estado,a)
        custo = p.path_cost(custo,estado,a,seg)
        estado = seg
    return (estado,custo)

#ALEATORIO FEITO GOSTO
def geradorAleatorio(dicionario):
    colar1 = [None] * 20
    colar2 = [None] * 20
    for i in range(20):
        index = random.randint(0,3)
        while dicionario[index] <= 0:
            index = random.randint(0,3)
        colar1[i] = index
        dicionario[index] -= 1
    for i in range(20):
        if i == 13:
            colar2[i] = colar1[7]
        elif i == 17:
            colar2[i] = colar1[3]
        else:
            index = random.randint(0,3)
            while dicionario[index] <= 0:
                index = random.randint(0,3)
            colar2[i] = index
            dicionario[index] -= 1
    return [colar1,colar2]
```
